```python
def convert_specifications(spec):
    my_dict = {}
    L1 = []
    L2 = []
    items = spec.split('|')
    for i, item in enumerate(items):
        if i % 2 == 0:
            L1.append(item.lower())
        else:
            L2.append(item.lower())
    return dict(zip(L1, L2))

# Apply the function to the 'specifications' column
flats['specifications'] = flats['specifications'].apply(lambda x: convert_specifications(x))
```

```python

# Function to expand the configuration column
def expand_configuration(df):
    # Initialize an empty list to store expanded rows
    rows = []
    # Iterate over each row in the DataFrame
    for _, row in df.iterrows():
        # Iterate over each item in the 'configuration' list
        for config in row['configuration']:
            # Create a copy of the row
            new_row = row.copy()
            # Update the 'configuration' column with the current configuration
            new_row['configuration'] = config
            # Append the modified row to the list
            rows.append(new_row)
    # Create a new DataFrame from the list of rows
    df_expanded = pd.DataFrame(rows)
    return df_expanded

# Expand the configuration column
flats_expanded = expand_configuration(flats)

# Reset index
flats_expanded.reset_index(drop=True, inplace=True)

# Display the expanded DataFrame
print(flats_expanded)
```

```python
def convert_config_column(fig):
    df = pd.DataFrame()
    for i, item in enumerate(fig):
        df1 = pd.DataFrame([item], columns=[f'column{k}' for k in range(len(item))], index=[i])
        df = pd.concat([df, df1])
    return df
```
