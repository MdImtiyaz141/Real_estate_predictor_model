```python
import numpy as np
import pandas as pd
import plotly.express as px
import seaborn as sns
```

```python
data = pd.read_csv('/content/before_model_df.csv')
data
```

```python
lat_long = pd.read_csv('/content/lat_long.csv')
```

```python
df = pd.merge(data,lat_long,left_on='property_name',right_on='property_name',how='inner')[['property_name','column5','column3','lat','long']].drop_duplicates()
```

```python
df
```

```python
cleaned_data = pd.read_csv('/content/Cleaned_data.csv')
```

```python
df[['column5','column3']] = cleaned_data[['avg_sba','Avg_prics']].values
```

```python
cleaned_data[['avg_sba','Avg_prics']]
```

```python
df.rename(columns={'column5':'SBA','column3':'avg_price'},inplace=True)
```

```python
df
```

```python
df.to_csv('chart_data.csv',index=False)
```

```python
df['lat']=pd.to_numeric(df['lat'],errors='coerce')
```

```python
df['long']=pd.to_numeric(df['long'],errors='coerce')
```

```python
df['column5'].isnull().sum()
```

```python
fig = px.scatter_mapbox(df,color='avg_price',lat="lat", lon="long",size='SBA',
                  color_continuous_scale=px.colors.cyclical.IceFire,zoom=10,
                  mapbox_style='open-street-map',text='property_name')
fig.show()
```
